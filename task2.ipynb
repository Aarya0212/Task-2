```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv('employee_dataset_for_eda.csv')
ds
```

```python
ds.info()
```

```python
#employee count as per department
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title('Employee count as per department')
```

Finance department having more employee count followed by marketing & engineering

```python
#gender distribution
sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title('gender distribution')
```

Female employees are more than male in specified companies

```python
#location distribution
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title('Number of employees by location')
plt.xticks(rotation=45)
```

mostly employees are located in seattle followed by san franciso & new york

```python
#age distribution
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title('Age distribution')
```

employee having age 45 to 50 are more than employee having age 25 to 30

```python
#Years_of_Experience distributiobn
sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title('Years_of_Experience')
```

Employee having more experience between range 2.5 to 5.0

```python
#monthly salary distribution
sns.boxplot(data=ds,y='Monthly_Salary',color='lightblue')
plt.title('monthly salary')
```

Each department has average salary in range 75,000

```python
#Average salary by department
sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette='viridis')
plt.title('Average salary by department')
```

Finance department has the average salary between 75000 to above 80000

```python
#performane scored by department
sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title('performane scored by department')
```

marketing department scored the maximum performance

```python
#salary distribution by gender
sns.violinplot(data=ds,x='Gender',y='Monthly_Salary',palette='muted')
plt.title('salary distribution by gender')
```

Male employee having maximum monthly salary

```python
#correlation matrix
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title('correlation matrix')
```

employee having less age and so that they also have leaa year of experience

```python
#salary vc experience scatter plot
sns.scatterplot(data=ds,x='Years_of_Experience',y='Monthly_Salary',hue='Department')
```
